```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Importing the data

df = pd.read_csv("C://Users//Snigdha.Cheekoty//OneDrive - Serco//Desktop//kaggle_ashrae//train.csv",
                parse_dates = True, index_col = "timestamp")
```

```python
df.index
```

```python
## Extracting only the date from the timestamp 

# Since the column values are already in timestamp format, we needn't convert them to datetime format and then exxtract date!
# The date is in format: %Y-%M-%D
import datetime
df.index = [datetime.datetime.date(d) for d in df.index]
```

```python
print(df.index)
df.index.name = "Date" ## naming the index with "Date"
```

```python
## Converting the datatype of index back to datetime
df.index = pd.to_datetime(df.index)
df.index
```

```python
## Grouping the data by date 
data = df.groupby(by = ["Date", "meter"]).sum()["meter_reading"] 
```

```python
data = data.to_frame() ## Converting the pandas Series to DataFrame
```

```python
data.head(20)
```

```python
data.tail(20)
```

```python
## Using a pivot table to spread out the data into different columns

data_pivot = pd.pivot_table(data, index = "Date", columns = "meter", values = "meter_reading")
data_pivot
```

```python
print(data_pivot.shape)
def color_entry(entry):

  if entry < 10000000:
    color = 'red'
  elif entry > 10000000:
    color = 'green'
  else:
    color = 'black'

  return 'color: %s' % color

data_pivot.style.applymap(color_entry).format("{:f}") 
## converting the scientific notation of the entries to floating point numbers, also changing colors 
##...... for better clarity 
## as you can see, this training data is of year 2016: for four different meters 
```
